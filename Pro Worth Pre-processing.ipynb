## Setting Up Datasets

### Grouping Market Value Per Year

```python
import pandas as pd

df = PlayerValuations
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year

years_valuations = pd.DataFrame(df['player_id'].unique(), columns=['player_id'])

for year in range(2018, 2025):
    yearly_data = df[df['year'] == year].groupby('player_id', as_index=False)['market_value_in_eur'].max()
    yearly_data.rename(columns={'market_value_in_eur': f'value_{year}'}, inplace=True)
    years_valuations = years_valuations.merge(yearly_data, on='player_id', how='left')

years_valuations.fillna(0, inplace=True)
```

### Setting Up Goals/Assits Dataset

#### Goals

```python
# Assuming your 'date' column is in a datetime format
PlayerPerformance['date'] = pd.to_datetime(PlayerPerformance['date'])  # Ensure the 'date' column is in datetime format

# Create a new 'year' column by extracting the year from 'date'
PlayerPerformance['year'] = PlayerPerformance['date'].dt.year

# Display the DataFrame to confirm the new column
PlayerPerformance.head()
```

```python
goals = PlayerPerformance[PlayerPerformance['type'] == 'Goals'][['year', 'player_id']]
goals
```

```python
# Initialize an empty DataFrame to store the results
playergoals = pd.DataFrame(goals['player_id'].unique(), columns=['player_id'])

# Loop through each year from 2014 to 2024
for year in range(2018, 2025):
    # Filter data for the specific year and count the number of rows per player_id (goals)
    yearly_data = goals[goals['year'] == year].groupby('player_id').size().reset_index(name='goals')
    
    # Rename the column for clarity
    yearly_data.rename(columns={'goals': f'goals_{year}'}, inplace=True)
    
    # Merge the yearly data into the goals_valuations DataFrame
    playergoals = playergoals.merge(yearly_data, on='player_id', how='left')

# Fill missing values with 0 (optional)
playergoals.fillna(0, inplace=True)

# Display the resulting DataFrame
playergoals.head()
```

#### Assists

```python
PlayerPerformance
```

```python
Assists = PlayerPerformance[PlayerPerformance['type'] == 'Goals'][['year','player_assist_id']]
Assists
```

```python
# Initialize an empty DataFrame to store the results
playerassists = pd.DataFrame(Assists['player_assist_id'].unique(), columns=['player_assist_id'])

# Loop through each year from 2014 to 2024
for year in range(2018, 2025):
    # Filter data for the specific year and count the number of rows per player_id (assists)
    yearly_data = Assists[Assists['year'] == year].groupby('player_assist_id').size().reset_index(name='assists')
    
    # Rename the column to reflect the year (e.g., 'year_2014_assists')
    yearly_data.rename(columns={'assists': f'assists_{year}'}, inplace=True)
    
    # Merge the yearly data into the playerassists DataFrame
    playerassists = playerassists.merge(yearly_data, on='player_assist_id', how='left')

# Fill missing values with 0 (optional)
playerassists.fillna(0, inplace=True)

# Display the resulting DataFrame
playerassists.head()
```

#### Appearances

```python
Appearances=pd.read_csv(r'D:\Football Project\Datasets\appearances.csv')
```

```python
# Convert 'date' column to datetime
Appearances['date'] = pd.to_datetime(Appearances['date'], errors='coerce')

# Now extract the year and group by it
Appearances['year'] = Appearances['date'].dt.year

# Group by year and count the rows
appearances_per_year = Appearances.groupby('year').size()

# Display the result
appearances_per_year
```

```python
# First, ensure the 'date' column is in datetime format and extract the year
Appearances['date'] = pd.to_datetime(Appearances['date'], errors='coerce')
Appearances['year'] = Appearances['date'].dt.year

# Filter for years between 2019 and 2024
filtered_appearances = Appearances[Appearances['year'].between(2019, 2024)]

# Group by player_id and year, sum minutes_played, and unstack
sum_minutes = filtered_appearances.groupby(['player_id', 'year'])['minutes_played'].sum().unstack(fill_value=0)

# Check for continuous zeros over 3 or more years
continuous_zero_filter = sum_minutes.apply(lambda row: any(row[i:i+2].eq(0).all() for i in range(len(row) - 1)), axis=1)

# Remove rows where there are 3 consecutive years with zero minutes
sum_minutes_cleaned = sum_minutes[~continuous_zero_filter]
sum_minutes_cleaned = sum_minutes_cleaned.rename(columns={col: f"minutes_{col}" for col in sum_minutes_cleaned.columns})

sum_minutes_cleaned
```

#### Injuries

```python
# Convert the 'from' column to datetime format
Injuries['from'] = pd.to_datetime(Injuries['from'], errors='coerce')

# Extract the year from the 'from' date
Injuries['year'] = Injuries['from'].dt.year

# Filter for years 2019 to 2022
Injuries_filtered = Injuries[Injuries['year'].between(2019, 2024)]

# Group by player_id and year, count injuries
injuries_per_year = Injuries_filtered.groupby(['player_id', 'year']).size().unstack(fill_value=0)

# Rename columns to 'injuries_<year>'
injuries_per_year.columns = [f'injuries_{col}' for col in injuries_per_year.columns]

# Display the result
injuries_per_year
```

## Merging Datasets

```python
#removing all zeros

years_valuations = years_valuations[(years_valuations != 0).all(axis=1)]
years_valuations
```

```python
import pandas as pd

# Ensure the column is in datetime format
Playerdetails['contract_expiration_date'] = pd.to_datetime(Playerdetails['contract_expiration_date'])

# Create a feature: Days until contract expiration
Playerdetails['days_until_expiration'] = (Playerdetails['contract_expiration_date'] - pd.Timestamp.today()).dt.days

# Replace NaN values with 0
Playerdetails['days_until_expiration'] = Playerdetails['days_until_expiration'].fillna(0)

# Keep only the necessary columns
selected_data = Playerdetails[['player_id', 'days_until_expiration']]

# Display the result
print(selected_data.head())
```

```python
import pandas as pd

# Assuming you have three DataFrames: years_valuations, playergoals, and playerassists
# Merging all three on 'player_id'

merged_df = pd.merge(years_valuations, playergoals, on='player_id', how='left')
merged_df = pd.merge(merged_df, playerassists, on='player_id', how='left')
merged_df = pd.merge(merged_df, selected_data, on='player_id', how='left')
merged_df = pd.merge(merged_df, sum_minutes_cleaned, on='player_id', how='left')
merged_df = pd.merge(merged_df, player_info, on='player_id', how='left')
merged_df = pd.merge(merged_df, injuries_per_year, on='player_id', how='left')




```

```python
merged_df.fillna(0, inplace=True)
merged_df
```

```python
# Display all column names of the merged DataFrame
print(merged_df.columns)
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Encode the 'country_of_citizenship' column
merged_df['country_of_citizenship_encoded'] = label_encoder.fit_transform(merged_df['country_of_citizenship'])

# Encode the 'sub_position' column
merged_df['sub_position_encoded'] = label_encoder.fit_transform(merged_df['sub_position'])


```
